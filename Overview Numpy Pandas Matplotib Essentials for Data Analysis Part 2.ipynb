```python
# Pie chart
import numpy as np
import matplotlib.pyplot as plt

#n = 20
#Z = np.random.uniform(0,1,n)
Z = np.array([175,50,225,150,400])
labs=["Java","C++","Python","PHP","Data Science"]
plt.title("Dummy graph", loc='left', size=12)
patch = plt.pie(Z, explode=(0, 0.9, 0, 0.1, 0), colors=['b','r','g','y','c'], radius=1.2, shadow=True, startangle=135,
        counterclock=False, autopct='%1.1f%%')
plt.legend(["Java","C++","Python","PHP","Data Science"], loc="upper right")
# plt.legend?
# plt.title?
plt.gca().set_aspect('equal')
plt.show() 
```
